```python
# @title Distribution of Rides per Week

import matplotlib.pyplot as plt
import pandas as pd

# Assuming your data is in a pandas DataFrame called 'df'
plt.figure(figsize=(10, 6))
for week in df['week'].unique():
    subset = df[df['week'] == week]
    plt.hist(subset['rides'], bins=10, alpha=0.5, label=f'Week {week}')
plt.xlabel('Number of Rides')
plt.ylabel('Frequency')
plt.title('Distribution of Rides per Week')
_ = plt.legend()
```

# Here we can see that after 2 to 3 rides there is a significant reduction in number of rides. This means that the average user is using anywhere between 2-3 rides in a week.
